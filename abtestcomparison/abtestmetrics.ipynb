## AB test metrics

```python
data_metrics = data_metrics.merge(data[['user_id','variant']], on=['user_id'], how='left')
data_metrics['variant'] = data_metrics['variant'].fillna('NotAssigned')
data_metrics[['user_id','dt','variant']]
```

```python
data_metrics_dau = data_metrics.groupby(['dt','variant']).agg(
    dau=('user_id', 'nunique')
).reset_index()

data_metrics_dau = data_metrics_dau[data_metrics_dau['variant'] != 'NotAssigned']

data_metrics_dau
```

### DAU

```python
# Calculate absolute and relative differences between variants
dau_pivot = data_metrics_dau.pivot(index='dt', columns='variant', values='dau')
dau_pivot['abs_diff'] = (dau_pivot['Test'] - dau_pivot['control']).abs()
dau_pivot['rel_diff_pct'] = ((dau_pivot['Test'] - dau_pivot['control']) / dau_pivot['control'] * 100).round(2)

# Merge back to original dataframe
data_metrics_dau = data_metrics_dau.merge(
    dau_pivot[['abs_diff', 'rel_diff_pct']], 
    left_on='dt', 
    right_index=True, 
    how='left'
)

data_metrics_dau
```

```python
import plotly.express as px

fig = px.line(data_metrics_dau, x='dt', y='dau', color='variant', 
              title='Daily Active Users by Variant',
              labels={'dt': 'Date', 'dau': 'DAU', 'variant': 'Variant'},
              markers=True,
              width=1200,
              height=600)

fig.add_vline(
    x=pd.Timestamp('2026-02-25').timestamp() * 1000, 
    line_dash='dash', 
    line_color='red', 
    annotation_text='Test Start', 
    annotation_position='top right')

fig.show()
```
