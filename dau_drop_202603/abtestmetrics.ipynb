```python
# Calculate absolute and relative differences between variants
dau_pivot = data_metrics_dau.pivot(index='dt', columns='variant', values='dau')
dau_pivot['abs_diff'] = (dau_pivot['Test'] - dau_pivot['control']).abs()
dau_pivot['rel_diff_pct'] = ((dau_pivot['Test'] - dau_pivot['control']) / dau_pivot['control'] * 100).round(2)

# Merge back to original dataframe
data_metrics_dau = data_metrics_dau.merge(
    dau_pivot[['abs_diff', 'rel_diff_pct']], 
    left_on='dt', 
    right_index=True, 
    how='left'
)

data_metrics_dau = data_metrics_dau[(data_metrics_dau['dt'] >= pd.Timestamp('2026-02-28'))]

data_metrics_dau
```

## Return rate

```python
return_rate_metrics = return_rate.merge(data[['user_id', 'variant']], on='user_id', how='left')
return_rate_metrics['variant'] = return_rate_metrics['variant'].fillna('NotAssigned')

return_rate_metrics['day_1'] = return_rate_metrics.loc[return_rate_metrics['day_1'], 'user_id']
return_rate_metrics['day_3'] = return_rate_metrics.loc[return_rate_metrics['day_3'], 'user_id']
return_rate_metrics['day_7'] = return_rate_metrics.loc[return_rate_metrics['day_7'], 'user_id']
return_rate_metrics['day_14'] = return_rate_metrics.loc[return_rate_metrics['day_14'], 'user_id']
return_rate_metrics['day_28'] = return_rate_metrics.loc[return_rate_metrics['day_28'], 'user_id']

return_rate_metrics = return_rate_metrics[(return_rate_metrics['dt'] >= pd.Timestamp('2026-02-28')) ]

return_rate_metrics = return_rate_metrics[return_rate_metrics['variant'] != 'NotAssigned']
```

```python
return_rate_metrics = return_rate_metrics.groupby(['dt', 'variant']).agg(
    total_users=('user_id', 'nunique'),
    rr_d1=('day_1', 'nunique'),
    rr_d3=('day_3', 'nunique'),
    rr_d7=('day_7', 'nunique'),
    rr_d14=('day_14', 'nunique'),
    rr_d28=('day_28', 'nunique'),
).reset_index()

return_rate_metrics['rr_d1_pct'] = return_rate_metrics['rr_d1'] / return_rate_metrics['total_users']
return_rate_metrics['rr_d3_pct'] = return_rate_metrics['rr_d3'] / return_rate_metrics['total_users']
return_rate_metrics['rr_d7_pct'] = return_rate_metrics['rr_d7'] / return_rate_metrics['total_users']
return_rate_metrics['rr_d14_pct'] = return_rate_metrics['rr_d14'] / return_rate_metrics['total_users']
return_rate_metrics['rr_d28_pct'] = return_rate_metrics['rr_d28'] / return_rate_metrics['total_users']



return_rate_metrics

```

```python
import plotly.graph_objects as go

# Prepare data for multiple return rate metrics
rr_metrics_melted = return_rate_metrics.melt(
    id_vars=['dt', 'variant'],
    value_vars=['rr_d1_pct', 'rr_d3_pct', 'rr_d7_pct', 'rr_d14_pct', 'rr_d28_pct'],
    var_name='metric',
    value_name='rate'
)

# Clean up metric names
rr_metrics_melted['metric'] = rr_metrics_melted['metric'].str.replace('rr_', '').str.replace('_pct', '')

# Filter out zero values
rr_metrics_melted = rr_metrics_melted[rr_metrics_melted['rate'] > 0]


```

```python
fig = go.Figure()

# Add traces for each variant-metric combination
for variant in return_rate_metrics['variant'].unique():
    for metric in ['d1', 'd3', 'd7', 'd14', 'd28']:
        data_subset = rr_metrics_melted[(rr_metrics_melted['variant'] == variant) & (rr_metrics_melted['metric'] == metric)]
        
        # Merge with return_rate_metrics to get total_users and rr values
        data_with_values = data_subset.merge(
            return_rate_metrics[['dt', 'variant', 'total_users', f'rr_{metric}']],
            on=['dt', 'variant'],
            how='left'
        )
        
        fig.add_trace(go.Scatter(
            x=data_with_values['dt'],
            y=data_with_values['rate'],
            mode='lines',
            name=f'{variant} - {metric}',
            hovertemplate=f'{variant} - {metric}<br>Date: %{{x}}<br>Rate: %{{y:.2%}}<br>Total Users: %{{customdata[0]}}<br>RR Count: %{{customdata[1]}}<extra></extra>',
            customdata=data_with_values[['total_users', f'rr_{metric}']].values
        ))

fig.update_layout(
    title='Return Rate by Day and Variant',
    xaxis_title='Date',
    yaxis_title='Return Rate',
    width=1200,
    height=600,
)

fig.add_vline(x=pd.Timestamp('2026-02-25').timestamp() * 1000, line_dash='dash', line_color='red', line_width=1, annotation_text='Test Start', annotation_position='top right')
fig.add_vline(x=pd.Timestamp('2026-03-13').timestamp() * 1000, line_dash='dash', line_color='red', line_width=1, annotation_text='Low period', annotation_position='top right')
fig.add_vline(x=pd.Timestamp('2026-03-25').timestamp() * 1000, line_dash='dash', line_color='red', line_width=1, annotation_text='Test End', annotation_position='top right')

fig.show()
```

## Player level

```python
player_level_metrics = player_level.merge(data[['user_id', 'variant']], on='user_id', how='left')
player_level_metrics['variant'] = player_level_metrics['variant'].fillna('NotAssigned')

player_level_metrics = player_level_metrics[(player_level_metrics['dt'] >= pd.Timestamp('2026-02-28')) & (player_level_metrics['dt'] <= pd.Timestamp('2026-03-24'))]

player_level_metrics['total_users_variant'] = player_level_metrics.groupby('variant')['user_id'].transform('nunique')


player_level_metrics = player_level_metrics.groupby(['max_level','variant','total_users_variant']).agg(
    user_count=('user_id', 'nunique')
).reset_index()

player_level_metrics.sort_values(['max_level','variant'], inplace=True)

player_level_metrics = player_level_metrics[player_level_metrics['variant'] != 'NotAssigned']

# Calculate percentage of user_count from total
player_level_metrics['pct_of_total'] = (player_level_metrics['user_count'] / player_level_metrics['total_users_variant'])

# Calculate relative difference between variants
# Pivot to get Test and control side by side
player_level_pivot = player_level_metrics.pivot(index='max_level', columns='variant', values='user_count').fillna(0)
player_level_pivot['rel_diff_pct'] = ((player_level_pivot['Test'] - player_level_pivot['control']) / player_level_pivot['control'])

# Merge back the relative difference
player_level_metrics = player_level_metrics.merge(
    player_level_pivot[['rel_diff_pct']], 
    left_on='max_level', 
    right_index=True, 
    how='left'
)

player_level_metrics
```
